```python
embeddings = []

for sure_index, sure_file_name in enumerate(sure_list):
    df = pd.read_csv(quran_prefix + sure_file_name)
    sure_name = sure_file_name.split('.')[0]
    for idex, idex_end in set([(0, i) for i in range(1, min(window_len, df.shape[0]))] + [(i, min(df.shape[0], i+window_len)) for i in range(0, df.shape[0])]):
        try:
            print(sure_name, idex, idex_end)
            # verse = row['verse']
            verse = ' '.join(df['ayahs'][i] for i in range(idex, idex_end))
            
            # Tokenize the verse
            inputs = tokenizer(verse, return_tensors="pt", padding=True, truncation=True)
            # print(inputs['input_ids'])
            
            # Get embeddings from the model
            with torch.no_grad():
                outputs = model(**inputs)
                # Use the last hidden state as embeddings (you can adjust this if needed)
                embedding = outputs.last_hidden_state[:, 0, :].squeeze().tolist()
            
            # Save the embedding with its corresponding index
            embeddings.append({
                "sure": sure_index,
                "sure_name": sure_name,
                "aye_start": idex+1,
                "aye_end": idex_end,
                "len": idex_end - idex,
                "window_len": window_len,
                "verse": verse,
                "embedding": embedding
            })
        except RuntimeError as e:
            print("Error in this: ", sure_name, idex, idex_end, e)
        
```

```python
import copy

# Save the embeddings to a file
# output_file = "quran_embeddings.json"
# with open(output_file, "w", encoding="utf-8") as f:
#     json.dump(embeddings, f, ensure_ascii=False, indent=4)

output_file = "quran_embeddings.csv.gz"
embeddings2 = copy.deepcopy(embeddings)
for e in embeddings2:
    e['embedding'] = ' '.join(str(x) for x in e['embedding'])
df = pd.DataFrame.from_dict(embeddings2)
df.to_csv(output_file, compression='gzip')

print(f"Embeddings saved to {output_file}")
```
